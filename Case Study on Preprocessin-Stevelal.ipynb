# 4. Fill in all the missing values present in all the columns in the dataset

```python
data.isnull().sum()
```

Columns Age, Cabin and Embarked are having null values

```python
# Handling the NaN in Age using median
data['Age']=data['Age'].fillna(data['Age'].median())

# Handling the Nan in Cabin setting it as unknown
data['Cabin']= data['Cabin'].fillna('Unknown')

# Handling the NaN in Embarked using mode
data['Embarked']= data['Embarked'].fillna(data['Embarked'].mode()[0])
```

# 5. Check and handle outliers in at least 3 columns in the dataset

```python
# Checking for outliers using boxplot
sns.boxplot(data['Age'])
```

```python
# Finding the outliers
Q1=np.percentile(data['Age'],25,interpolation='midpoint')
Q2=np.percentile(data['Age'],50,interpolation='midpoint')
Q3=np.percentile(data['Age'],75,interpolation='midpoint')
print("The values of Q1, Q2, Q3 are:",'\n', Q1, Q2, Q3)
IQR=Q3-Q1
print("IQR value= ",IQR)
upp_limit=Q3+1.5*IQR
low_limit=Q1-1.5*IQR
print('UPPER LIMIT = ',upp_limit)
print('LOWER LIMIT = ',low_limit)
outlier=[]
for i in data['Age']:
    if((i>upp_limit) | (i<low_limit)):
        outlier.append(i)
print("outlier = ",outlier)
```

```python
# Using standard scale to rescale the outlier value
from sklearn.preprocessing import StandardScaler
col=['Age']
feature=data[col]
scaler=StandardScaler().fit(feature.values)
feature=scaler.transform(feature.values)
data[col]=feature
data
```

# 6. Do min max scaling on the feature set (Take ‘Survived’ as target)

```python
from sklearn import preprocessing
minmax_scale = preprocessing.MinMaxScaler().fit(data[['Survived','Age']])
data_minmax = minmax_scale.transform(data[['Survived', 'Age']])
```

```python
print('Min-value after min-max scaling:\nSurvived={:.2f}, Age={:.2f}'
      .format(data_minmax[:,0].min(), data_minmax[:,1].min()))
print('\nMax-value after min-max scaling:\nSurvived={:.2f}, Age={:.2f}'
      .format(data_minmax[:,0].max(), data_minmax[:,1].max()))
```

```python
def plot():
    plt.figure(figsize=(8,6))

    plt.scatter(data['Survived'], data['Age'],
            color='green', label='input scale', alpha=0.5)

    plt.scatter(data_minmax[:,0], data_minmax[:,1],
            color='blue', label='min-max scaled [min=0, max=1]', alpha=0.3)

    plt.title('Survived and Age')
    plt.xlabel('Survived')
    plt.ylabel('Age')
    plt.legend(loc='upper left')
    plt.grid()

    plt.tight_layout()

plot()
plt.show()
```
