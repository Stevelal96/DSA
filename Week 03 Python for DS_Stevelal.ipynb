### 3. Select the rows the age is between 24 and 29 (inclusive)

### 4. Find the minimum and maximum value of insulin level in the data.

```python
Dataset['Insulin'].max()
```

```python
Dataset['Insulin'].min()
```

### 5. Create a histogram of "Age".

```python
plt.figure(figsize= (5,5))
plt.hist(Dataset['Age'], color ='Black')
plt.title('Histogram of Age',fontsize=14)
plt.xlabel('Age',fontsize=12)
plt.ylabel('No of people',fontsize=12)
```

### 6. Create a scatter plot of "BMI" vs "Glucose".

```python
plt.figure(figsize=(5,5))
plt.scatter(Dataset['BMI'],Dataset['Glucose'],marker='^',color='black')
plt.title('Scatterplot of BMI vs Glucose',fontsize=14)
plt.xlabel('BMI',fontsize=12)
plt.ylabel('Glucose',fontsize=12)
```

### 7. Replace the min value of "BMI" with mean value of BMI

```python
a=Dataset['BMI'].mean()
a=round(a,2)
b=Dataset['BMI'].min()
Dataset['BMI']=Dataset['BMI'].replace(b,a)
new_min=Dataset['BMI'].min()
new_min
```

### 8. Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome

```python
sns.boxplot(x='Outcome',y='Glucose',data=Dataset)
```
